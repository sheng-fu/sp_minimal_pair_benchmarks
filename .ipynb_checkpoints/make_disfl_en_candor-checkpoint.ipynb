```python
import pandas as pd
from tqdm import tqdm
import re
import os
import random
random.seed(1991)
```

```python
candor_path = "C:/Users/Sheng-Fu/Documents/candor/"
```

```python
dirs = os.listdir(candor_path)
len(dirs)
```

```python
cand_utts = []
for i in dirs[:50]:
    cliffhanger = pd.read_csv(os.path.join(candor_path, i, "transcription/transcript_cliffhanger.csv"))
    utts = cliffhanger['utterance'].tolist()
    utts = [re.sub(r"(\w)([^\w\s'])", r"\1 #", x) for x in utts]
    utts = [x.split() for x in utts if 'uh huh' not in x.lower() and 'uh mhm' not in x.lower()]
    utts = [x for x in utts if len(x) > 4 and len(x) < 31 and ('uh' in x or 'Uh' in x or 'um' in x or 'Um' in x)]
    cand_utts = cand_utts + utts
    print(len(cand_utts))
```

```python
cand_utts[:10]
```

```python
random.shuffle(cand_utts)
```

```python
with open('english_fp_base.txt', 'w', encoding = 'utf-8') as f:
    for i in cand_utts:
        f.write(' '.join(i[:-1]) + '\n')
```

```python
cand_utts = []
for i in dirs[50:60]:
    cliffhanger = pd.read_csv(os.path.join(candor_path, i, "transcription/transcript_cliffhanger.csv"))
    utts = cliffhanger['utterance'].tolist()
    utts = [re.sub(r"(\w)([^\w\s'])", r"\1 #", x) for x in utts]
    utts = [x.split() for x in utts if x.count('#') > 1]
    utts = [x for x in utts if len(x) > 4 and len(x) < 31]
    cand_utts = cand_utts + utts
    print(len(cand_utts))
cand_utts[:10]
```

```python
random.shuffle(cand_utts)


with open('english_sp_base.txt', 'w', encoding = 'utf-8') as f:
    for i in cand_utts:
        f.write(' '.join(i[:-1]) + '\n')
```

```python
cand_utts = []
for i in dirs[60:200]:
    cliffhanger = pd.read_csv(os.path.join(candor_path, i, "transcription/transcript_cliffhanger.csv"))
    utts = cliffhanger['utterance'].tolist()
    utts = [re.sub(r"(\w)([^\w\s'])", r"\1 #", x).split() for x in utts]
    utts = [x for x in utts if len(x) > 4 and len(x) < 31]
    cand_utts = cand_utts + utts

repetition_pool = []
for c in cand_utts:
    temp = [x for x in c]
    for j in range(0,len(temp)-1):
        if temp[j] not in ['#', 'uh', 'um', 'Uh', 'Um'] and temp[j] == temp[j+1]:
            repetition_pool.append(c)
            break
```

```python
len(repetition_pool)

random.shuffle(repetition_pool)
```

```python
with open('english_repeats_base.txt', 'w', encoding = 'utf-8') as f:
    for i in repetition_pool:
        f.write(' '.join(i[:-1]) + '\n')
```
